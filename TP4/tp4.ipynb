### Ejercicio adicional

La práctica se llevó a cabo usando un monitor LED marca Samsung modelo LS22D300FYCZB, el cual contaba con 3 modos distintos para el Gamma: Modo1, Modo2 y Modo3. Se realizaron las mediciones empleando el firmware provisto junto con los códigos en Python, tanto para generar los distintos niveles de gris como para registrar las mediciones con el sensor y hacer el ajuste. Estos últimos se modificaron levemente, a modo de incluir el desvío estándar de la media de las mediciones como medida del error en las mediciones de luminancia, para que este figure en los gráficos y sea tenido en cuenta en el ajuste no lineal. Las mediciones obtenidas son resultado del promedio de 5 mediciones, y las barras de error representan el desvío estándar de la media.

```python
import cv2
from matplotlib import pyplot as plt

img1 = cv2.imread('A 1/Modo 1.png')
img2 = cv2.imread('A 1/Modo 2.png')
img3 = cv2.imread('A 1/Modo 3.png')

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img1)
plt.title('Modo 1')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img2)
plt.title('Modo 2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img3)
plt.title('Modo 3')
plt.axis('off')

plt.tight_layout()
plt.show()
```

Se observa que el gamma aumenta con el número de modo, resultando en curvas más pronunciadas donde prevalencen tonos más oscuros (de menor luminancia) en el monitor.

Cabe destacar que en los 3 casos se observa que no es posible resolver los últimos dos niveles de entrada, los cuales resultan en luminancias idénticas. Creemos que esto no se debe a una saturación del sensor, sino a una saturación del monitor, ya que incluso a simple vista no era posible observar diferencias entre estos últimos dos niveles.

Para solucionar este problema, se re movió el último valor de los niveles de entrada y se volvieron a realizar los ajustes:

```python
img1 = cv2.imread('A 09/Modo 1.png')
img2 = cv2.imread('A 09/Modo 2.png')
img3 = cv2.imread('A 09/Modo 3.png')

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img1)
plt.title('Modo 1')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img2)
plt.title('Modo 2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img3)
plt.title('Modo 3')
plt.axis('off')

plt.tight_layout()
plt.show()
```

Se observa ahora que los valores de gamma, si bien mantienen la misma tendencia, se modificaron notablemente. Para el Modo 1, pareciera que los datos no captan una tendencia lo suficientemente exponencial como para que se pueda realizar el ajuste, obteniéndose un valor de gamma igual a 1.
